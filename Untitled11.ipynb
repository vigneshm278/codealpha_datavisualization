```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
v=pd.read_csv("/content/Telco-Customer-Churn.csv")
v.head()
```

```python
print("shape:", v.shape)
print("Data Types:\n", v.dtypes)
v.head()
```

```python
v.describe()
```

```python
v.info()
```

```python
def con(value):
    if value == 1:
        return "yes"
    else:
        return "no"

v["SeniorCitizen"] = v["SeniorCitizen"].apply(con)
print(v.columns)

```

```python
v.head(20)
```

```python
ax = sns.countplot(x="Churn", data=v)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(6, 4))
sns.countplot(x="Churn", data=v, palette="Set2")
plt.title("Churn Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="gender", hue="Churn", data=v , palette="pastel")
plt.title("churn by Gender")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x="Contract", hue="Churn", data=v , palette="cool")
plt.title("churn by Contract Type")
plt.xticks(rotation=15)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(x="MonthlyCharges", hue="Churn", fill=True, palette="viridis", data=v)
plt.title("Monthly Charges Distribution by Churn")
plt.show()
```

```python
v['TotalCharges'] = pd.to_numeric(v["TotalCharges"], errors='coerce')
plt.figure(figsize=(10, 6))
sns.heatmap(v.corr(numeric_only=True), annot=True, cmap="Blues")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize= (3,4))
gb = v.groupby("Churn").agg({'Churn': "count"})
plt.pie(gb["Churn"], labels=gb.index, autopct="%1.2f%%")
plt.title("Percentage of Churn Customers")
plt.show()
```

```python
plt.figure(figsize= (9, 4))
sns.histplot(x = "tenure", data = v, bins = 50 , hue = "Churn")
plt.show()
```

```python
v.columns.values
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

sns.set(style="whitegrid")
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(cols):
    ax = axes[i]
    sns.countplot(data=v, x=col, ax=ax, palette='Set2', hue="Churn")  # ✅ use column name, not df["Churn"]
    ax.set_title(f'Count of {col}')
    ax.tick_params(axis='x', rotation=20)

    if hasattr(ax, 'bar_label'):
        for container in ax.containers:
            ax.bar_label(container, label_type='edge')

    ax.grid(False)

for j in range(len(cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
